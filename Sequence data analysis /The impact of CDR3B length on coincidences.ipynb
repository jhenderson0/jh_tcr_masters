# An analysis into the effect of length on the coincidence statistics of TCRB chains

## Module imports and figure formatting

```python
import pandas as pd
import numpy as np

#Pyrepseq imports
import pyrepseq as prs
import pyrepseq.plotting as rsp
import pyrepseq.distance as rsd
import pyrepseq.stats as rss

#Plotting imports
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm
import matplotlib.style 
import matplotlib as mpl 
from matplotlib.colors import LogNorm, Normalize
import seaborn as sns

from scipy.optimize import curve_fit
```

```python
#Set default plotting parameters
mpl.rcParams['figure.figsize'] = [4.0, 4.0] 
mpl.rcParams["legend.frameon"] = False
mpl.rcParams['figure.dpi']=500 

SMALL_SIZE = 10
MEDIUM_SIZE = 15
BIGGER_SIZE = 15

plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

## Data imports 

### Mira - specific set

```python
mira = pd.read_csv('../data/mira/processed/peptide-detail-ci.csv')
mira = mira.drop_duplicates("TCR Nucleotide Sequence")
mira = prs.standardize_dataframe(mira
                                        , [None, None, None ,
                               "v_gene", "amino_acid", "j_gene",
                    
                                           "Amino Acids", None, None, None])

mira = mira.dropna(axis=0, subset = ["CDR3B"] ).reset_index().drop("index", axis=1)
```

### Minervina - background

```python
back = pd.read_csv('../data/minervina/beta/W_F1_2019_beta.txt.gz',sep='\t',index_col=0)
back = back.drop_duplicates("nSeqCDR3")
back = prs.standardize_dataframe(back
                                        , [None, None, None ,
                               "bestVGene", "aaSeqCDR3", "bestJGene",
                    
                                           None, None, None, None])
```

## Length calculation

```python
mira["CDR3B length"] = mira["CDR3B"].apply(len)-2 #Take 2 for the C and the F

back["CDR3B length"] = back["CDR3B"].apply(len)-2

unique_lengths_mira = pd.unique(mira["CDR3B length"])
unique_lengths_back = pd.unique(back["CDR3B length"])

min_length = np.amin(unique_lengths_back)
max_length = np.max(unique_lengths_back)

bin_edges = np.arange(min_length-0.5, max_length+2.5)
```

## Length distribution 

### Across epitopes

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.hist(mira["CDR3B length"], bins = bin_edges, edgecolor='black'
        , density = True, label='mira')
ax.hist(back["CDR3B length"], bins = bin_edges, edgecolor='black'
        , density = True, label= 'minervina', alpha = 0.5)
ax.set_xlabel("CDR3B Length")
ax.set_ylabel("Density")
ax.legend();
```

### Within epitopes

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.hist(mira.groupby("Epitope")["CDR3B length"].mean(), edgecolor='black', density = True, bins = 25)
ax.set_xlabel("Mean length")
ax.set_ylabel("Density");
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.hist(mira.groupby("Epitope")["CDR3B length"].std(), edgecolor='black', density = True, bins = 25)
ax.set_xlabel("Standard deviation in length")
ax.set_ylabel("Density");
```

## Probability of length coincidence

```python
pc_length_back = prs.pc(back["CDR3B length"])
pc_length_spc = mira.groupby('Epitope').apply(lambda x: prs.pc(x["CDR3B length"])).mean()
print(f"CDR3B length coincidence ratio: {pc_length_spc/pc_length_back}")
```

## Fraction of specific sequences  and value at $\Delta = 1$ with length

```python
def linear_fit(x, a, b):
    
    return a*x + b
```

```python
def exp_fit(x, a, b):
    
    return a*np.exp(-b*x)
```

```python
min_l = 8
max_l = 14

fig,ax = plt.subplots(1,2, figsize=(10,5))

f_spc = 1/ratio.iloc[0][min_l-5:max_l-4]
lengths = np.arange(min_l, max_l+1)

length_filter = ~np.isin(lengths, [15])

f_spc_fit = curve_fit(exp_fit, lengths[length_filter], f_spc[length_filter])[0]
ax[0].plot(f_spc, marker = '.',ls='none')
ax[0].set_xlabel("CDR3B length")
ax[0].set_ylabel('Exact coincidence ratio')
ax[0].plot(lengths, exp_fit(lengths,*f_spc_fit))
ax[0].set_yscale("log")

delta_1_fit = curve_fit(linear_fit, lengths[length_filter], (ratio.iloc[1][min_l-5:max_l-4]*f_spc)[length_filter])[0]
ax[1].plot(ratio.iloc[1][min_l-5:max_l-4]*f_spc, marker = '.', ls='none')
ax[1].set_xlabel("CDR3B length")
ax[1].set_ylabel(r'$\left< p_s | \Delta = 1 \right>$')
ax[1].plot(lengths, linear_fit(lengths,*delta_1_fit))
#ax[1].set_yscale("log")

fig.tight_layout();
```
