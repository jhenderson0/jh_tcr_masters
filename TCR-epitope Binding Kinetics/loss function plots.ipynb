# Maximum likelihood estimate

## Imports and formatting

```python
import pandas as pd
import numpy as np

#Plotting imports
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm
import matplotlib.style 
import matplotlib as mpl 
import seaborn as sns
```

```python
#Set default plotting parameters
mpl.rcParams['figure.figsize'] = [4.0, 4.0] 
mpl.rcParams["legend.frameon"] = False
mpl.rcParams['figure.dpi']=500 

SMALL_SIZE = 10
MEDIUM_SIZE = 12
BIGGER_SIZE = 12

plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
def mle(x, E_star):
    
    return np.log(1+np.exp(x-E_star)) 
```

## Plot

```python
fig,ax = plt.subplots(figsize=(10,5))

xs = np.linspace(-20,20,1000)
E_stars = np.linspace(-5,5,5)

for E_star in E_stars:
    ax.plot(xs, mle(xs, E_star),label = f"$E^*$ = {E_star:.1f}")
ax.set_xlabel(r"$\Delta E$")
ax.set_ylabel(r"$\ell$")
ax.set_ylim([0,None])
ax.legend()
plt.savefig('loss-fun.png');
```
